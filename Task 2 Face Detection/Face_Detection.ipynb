```python
from google.colab import files
uploaded = files.upload()

"""
Task 2: Face Detection using OpenCV Haar Cascades
Kodbud AI Internship Program

This script detects human faces in multiple images
using OpenCV's pre-trained Haar Cascade classifier.
"""

# ==============================
# 1️⃣ Import Required Libraries
# ==============================

import cv2
from google.colab.patches import cv2_imshow


# ==========================================
# 2️⃣ Load Pre-trained Haar Cascade Model
# ==========================================

# OpenCV provides built-in Haar Cascade models
# for detecting faces, eyes, smiles, etc.

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)


# ==================================
# 3️⃣ List of Input Image Files
# ==================================

# Add all your images here
image_list = ['img.webp', 'img2.webp', 'img3.webp', 'img4.webp']

print(f"\n--- Starting Face Detection for {len(image_list)} images ---")


# ==================================
# 4️⃣ Process Each Image
# ==================================

for image_path in image_list:

    # 🔹 Read image
    img = cv2.imread(image_path)

    # Check if image loaded properly
    if img is None:
        print(f"❌ Error: Could not load {image_path}")
        continue

    # 🔹 Convert to Grayscale
    # Haar Cascades work better on grayscale images
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 🔹 Detect Faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,      # Detect faces at different scales
        minNeighbors=5       # Higher value = fewer false positives
    )

    # 🔹 Draw Rectangles Around Faces
    for (x, y, w, h) in faces:
        cv2.rectangle(
            img,
            (x, y),
            (x + w, y + h),
            (0, 255, 0),  # Green color (BGR)
            3             # Thickness
        )

    # ==================================
    # 5️⃣ Display Results
    # ==================================

    print(f"\n✅ Result for {image_path}: Found {len(faces)} face(s)")
    cv2_imshow(img)

print("\n--- Face Detection Completed ---")
```
